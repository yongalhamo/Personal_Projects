```python
Title=[]
Type =[]
Episodes=[]
Scores=[]
Members=[]
Favorites=[]
Genre=[]
Studio=[]
Rating=[]
Demographic=[]
Theme=[]
Popularity=[]
Source=[]
Rank=[]
limit=0

while limit !=4000:
    url = get(f"https://myanimelist.net/topanime.php?limit={limit}", headers=headers)
    request = url.text
    soup = bs(request,'html.parser')
    title = soup.findAll('tr',{'class':"ranking-list"})
    print(limit)
   
    for i in title:
        
        name = i.h3.text 
        Title.append(name)
        
        
        form = i.find('div', class_='information di-ib mt4').text.split()[0]
        Type.append(form)
        
        eps=i.find('div', class_='information di-ib mt4').text.split()[1].replace('(','')
        Episodes.append(eps)
        
        score=i.find('div',class_='js-top-ranking-score-col di-ib al').text
        Scores.append(score)
    
        
        
        a=i.find('h3',class_='hoverinfo_trigger fl-l fs14 fw-b anime_ranking_h3').a['href']
        link = get(a)
        request = link.text
        cont = bs(request,'html.parser')
    
        
        memb=cont.find('div',class_='di-ib ml12 pl20 pt8').find('span',{'class':'numbers members'}).text.split()[1]
        Members.append(memb)
        
        fav=cont.find('span','dark_text',text='Favorites:').next_sibling.strip()
        Favorites.append(fav)
        
        genre=cont.find('span',itemprop='genre').text
        Genre.append(genre)
        
        studio=cont.find('span',class_='dark_text',text='Studios:').next_sibling.next_sibling.text
        Studio.append(studio)
        
        rating=cont.find('span',class_='dark_text',text='Rating:').next_sibling.strip()
        Rating.append(rating)
        
        popularity=cont.find('span',class_='dark_text',text='Popularity:').next_sibling.strip()
        Popularity.append(popularity)
        
        source=cont.find('span',class_='dark_text',text='Source:').next_sibling.strip()
        Source.append(source)
        
        rank = cont.find('span',class_='numbers ranked' ).text.split()[1]
        Rank.append(rank)
        
        try:
            demo=cont.find('span',class_='dark_text',text='Demographic:').next_sibling.next_sibling.text
            Demographic.append(demo)
        except:
            Demographic.append("N/A")
            pass
        
        try:
            theme=cont.find('span',class_='dark_text',text='Themes:').next_sibling.next_sibling.text
            Theme.append(theme)
        except:
            
            try:
                theme=cont.find('span',class_='dark_text',text='Theme:').next_sibling.next_sibling.text
                Theme.append(theme)
            
            except:
                Theme.append("N/A")
         
       
        
    limit = limit+200  
    
    
    
```
